# Day 4

```python
# Open the input file
inputfile = open('day4_input.txt', 'r')

# Parse lines
data = [x.strip() for x in inputfile.readlines()]
```

```python
data
```

```python
req_fields = ['byr', 'iyr', 'eyr', 'hgt', 'hcl', 'ecl', 'pid']
crrnt = 0
valid_count = 0
byr = [None]*253
iyr = [None]*253
eyr = [None]*253
hgt = [None]*253
hcl = [None]*253
ecl = [None]*253
pid = [None]*253
cid = [None]*253

idx = 0
for line in data:
    if line == '':
        idx +=1
        if crrnt == len(req_fields):
            valid_count +=1
        crrnt = 0
        continue
    
    for field in line.split():
        key, val = field.split(':')
        if key in req_fields:
            crrnt += 1
        
            # create dataframe vectors
        if key == 'byr':
            byr[idx] = val
        if key == 'iyr':
            iyr[idx] = val
        if key == 'eyr':
            eyr[idx] = val
        if key == 'hgt':
            hgt[idx] = val
        if key == 'hcl':
            hcl[idx] = val
        if key == 'ecl':
            ecl[idx] = val
        if key == 'pid':
            pid[idx] = val
        if key == 'cid':
            cid[idx] = val
```

```python
comb_dict = {'byr':byr, 'iyr':iyr, 'eyr':eyr, 'hgt':hgt, 'hcl':hcl,
             'ecl':ecl, 'pid':pid, 'cid':cid}
pdf = pd.DataFrame(comb_dict)
```

```python
pdf
```

```python
pdf = pdf.fillna({'byr':'0', 'iyr':'0', 'eyr':'0', 'hgt':'0cm', 'hcl':'none',
             'ecl':'none', 'pid':pid, 'cid':cid})
```

```python
pdf2 = pdf.copy()
pdf2 = pdf2.drop('cid', axis=1)
pdf2 = pdf2.dropna()
```

```python
pdf2 = pdf2.reset_index()
```

```python
pdf2['byr'] = pdf2['byr'].astype(int)
pdf2['iyr'] = pdf2['iyr'].astype(int)
pdf2['eyr'] = pdf2['eyr'].astype(int)
#pdf2['pid'] = pdf2['pid'].astype(int)
```

```python
# lets impose filters in turn
# birth year, issue year and expiry year
pdf3 = pdf2[(pdf2.byr >= 1920) & (pdf2.byr <= 2020) &
            (pdf2.iyr >= 2010) & (pdf2.iyr <= 2020) &
            (pdf2.eyr >= 2020) & (pdf2.eyr <= 2030)]
pdf3.shape
```

```python
pdf3 = pdf3.drop('index', axis=1)
pdf3 = pdf3.reset_index()
pdf3 = pdf3.drop('index', axis=1)

# filter on valid height
pdf3['unit'] = pdf3['hgt'].str.extract('(\D+)') 

pdf3['hgt'] = pdf3['hgt'].str.extract('(\d+)').astype(int)

pdf4 = pdf3[((pdf3.unit == 'cm') & (pdf3.hgt >= 150) & (pdf3.hgt <= 193))|
            (( pdf3.unit == 'in') & (pdf3.hgt >= 59) & (pdf3.hgt <= 76))]
pdf4.shape
```

```python
# valid eye colour
pdf5 = pdf4[pdf4.ecl.isin(['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth'])]
pdf5.shape
```

```python
# Passport id
pdf6 = pdf5.copy()
pdf6['act_pid'] = pdf6['pid'].str.extract('^(\d{9})$', expand=False)
pdf6 = pdf6[pdf6.act_pid.notna()]
pdf6.shape
```

```python
def check_number(nme, contents_list, no):
    sum_no = no
    for c in contents_list:
        d = c.split(' ', 2)[2:][0]
        bag_type = c.split(' ', 2)[0][:]+ ' ' +c.split(' ', 2)[1][:]
        g = re.findall(nme, bag_type)
        if g:
            words_line = c.split(' ')
            for i, a in enumerate(words_line):
                if a.isdigit():
                    new_bag = words_line[i+1] + ' ' + words_line[i+2]
                    #print(a)
                    sum_no += int(a) + int(a)*(check_number(new_bag, contents_list, 0))
                elif (a == 'no'):
                    sum_no = 0
    return sum_no

# call function here
```
